1. Implement Finite Field Arithmetic

```python
def ffAdd(a, b):
    ans = a ^ b
    return ans
```

```python
def xtime(f):
    ans = f << 1
    if f & 0x80:
        m = 0x011b
        ans ^= m
        
    return ans
```

```python
def ffMultiply(a, b):
    ans = 0b00000000
    for i in range(8):
        if b & 1: #b's last digit is 1
#             print(str(i) + ": " + '\n' + "ans: " + str(ans) + '\n' + "a: " + str(a) + '\n')
            ans = ffAdd(ans, a)
        a = xtime(a) #increase powers of a by 1
        b >>= 1 #focus on next power of b
    
    return ans        
```

2. Implement Key Expansion

```python
def subWord(w):
    wString = hex(w)
    wString = wString[2:]
    wString = wString.zfill(8)
    wordList = []
    for i in range(4):
        wordList.append(wString[2*i:2*i+2])
    wordList

    ansList = []
    for i in range(4):
        ansList.append(Sbox[int(wordList[i], 16)])
    ansString = "0x" + hex(ansList[0])[2:].zfill(2) + hex(ansList[1])[2:].zfill(2) + hex(ansList[2])[2:].zfill(2) + hex(ansList[3])[2:].zfill(2)
    ans = int(ansString, 16)
     
    return ans
```

```python
def rotWord(w):
    wString = hex(w)
    wString = wString[2:]
    wString = wString.zfill(8)
    wordList = []
    
    for i in range(4):
        wordList.append(wString[2*i:2*i+2])
    ansString = "0x" + wordList[1] + wordList[2] + wordList[3] + wordList[0]
    ans = int(ansString, 16)
    
    return ans
```

```python
# 128 bit key
def keyExpansion128(key):
    w = []
    i = 0
    while i < 4:
        wString = "0x" + hex(key[4*i])[2:].zfill(2) + hex(key[4*i+1])[2:].zfill(2) + hex(key[4*i+2])[2:].zfill(2) + hex(key[4*i+3])[2:].zfill(2)
        w.append(int(wString, 16))
        i += 1

    i = 4
    while i < 4*(10+1):
        temp = w[i-1]
        if i%4 == 0:
            temp = subWord(rotWord(temp)) ^ Rcon[int(i/4)]

        w.append(w[i-4] ^ temp)
        i += 1
        
    return w
```

```python
# 192 bit key
def keyExpansion192(key):
    w = []
    i = 0
    while i < 6:
        wString = "0x" + hex(key[4*i])[2:].zfill(2) + hex(key[4*i+1])[2:].zfill(2) + hex(key[4*i+2])[2:].zfill(2) + hex(key[4*i+3])[2:].zfill(2)
        w.append(int(wString, 16))
        i += 1

    i = 6
    while i < 4*(12+1):
        temp = w[i-1]
        if i%6 == 0:
            temp = subWord(rotWord(temp)) ^ Rcon[int(i/6)]

        w.append(w[i-6] ^ temp)
        i += 1
        
    return w
```

```python
# 256 bit key
def keyExpansion256(key):
    w = []
    i = 0
    while i < 8:
        wString = "0x" + hex(key[4*i])[2:].zfill(2) + hex(key[4*i+1])[2:].zfill(2) + hex(key[4*i+2])[2:].zfill(2) + hex(key[4*i+3])[2:].zfill(2)
        w.append(int(wString, 16))
        i += 1

    i = 8
    while i < 4*(14+1):
        temp = w[i-1]
        if i%8 == 0:
            temp = subWord(rotWord(temp)) ^ Rcon[int(i/8)]
        elif i%8 == 4:
            temp = subWord(temp)

        w.append(w[i-8] ^ temp)
        i += 1
        
    return w
```

3. Implement the cipher function

```python
def subBytes(s):
    for c in range(4):
        for r in range(4):
            s[r][c] = Sbox[s[r][c]]       
    return s
```

```python
def shiftRows(s):
    ans = [[0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00]
    ]   
    for c in range(4):
        for r in range(4):
            ans[r][c] = s[r][(c + r) % 4]
    return ans
```

```python
def mixColumns(s):
    a = [[0x02, 0x03, 0x01, 0x01],
         [0x01, 0x02, 0x03, 0x01],
         [0x01, 0x01, 0x02, 0x03],
         [0x03, 0x01, 0x01, 0x02]
    ]
    ans = [[0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00]
    ]  
    
    for c in range(4):
        for r in range(4):
            ans[r][c] = ffMultiply(a[r][0], s[0][c]) ^ ffMultiply(a[r][1], s[1][c]) ^ ffMultiply(a[r][2], s[2][c]) ^ ffMultiply(a[r][3], s[3][c])
    return ans
```

```python
def addRoundKey(s, w):
    for c in range(4):
        roundKey = hex(w[c])[2:].zfill(8)
        for r in range(4):
            roundKeyPiece = roundKey[2*r:2*r+2]
            roundKeyPiece = int(roundKeyPiece, 16)
            s[r][c] ^= roundKeyPiece
    return s
```

```python
def cipher(i, w, numRounds):
    s = i
    print(transform(s))

    s = addRoundKey(s, w[0:4])
    print(transform(s))
    
    for r in range(1, numRounds):
        s = subBytes(s)
        print(transform(s))
        s = shiftRows(s)
        print(transform(s))
        s = mixColumns(s)
        print(transform(s))
        s = addRoundKey(s, w[r*4:(r+1)*4])
        print(transform(s))
    
    s = subBytes(s)
    print(transform(s))
    s = shiftRows(s)
    print(transform(s))
    s = addRoundKey(s, w[numRounds*4:(numRounds+1)*4])
    
    return s
```

4. Implement the invCipher function

```python
def invSubBytes(s):
    for c in range(4):
        for r in range(4):
            s[r][c] = InvSbox[s[r][c]]       
    return s
```

```python
def invShiftRows(s):
    ans = [[0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00]
    ]   
    for c in range(4):
        for r in range(4):
            ans[r][c] = s[r][(c - r) % 4]
    return ans
```

```python
def invMixColumns(s):
    a = [[0x0e, 0x0b, 0x0d, 0x09],
         [0x09, 0x0e, 0x0b, 0x0d],
         [0x0d, 0x09, 0x0e, 0x0b],
         [0x0b, 0x0d, 0x09, 0x0e]
    ]
    ans = [[0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00]
    ]  
    
    for c in range(4):
        for r in range(4):
            ans[r][c] = ffMultiply(a[r][0], s[0][c]) ^ ffMultiply(a[r][1], s[1][c]) ^ ffMultiply(a[r][2], s[2][c]) ^ ffMultiply(a[r][3], s[3][c])
    return ans    
```

```python
def invCipher(i, w, numRounds):
    s = i
    print(transform(s))

    s = addRoundKey(s, w[numRounds*4:(numRounds+1)*4])
    print(transform(s))
    
    for r in range(numRounds-1, 0, -1):
        s = invShiftRows(s)
        print(transform(s))
        s = invSubBytes(s)
        print(transform(s))
        s = addRoundKey(s, w[r*4:(r+1)*4])
        print(transform(s))
        s = invMixColumns(s)
        print(transform(s))
    
    s = invShiftRows(s)
    print(transform(s))
    s = invSubBytes(s)
    print(transform(s))
    s = addRoundKey(s, w[0:4])
    
    return s
```

```python
def convertPlainText128(t):    
    ans = [[0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00],
           [0x00, 0x00, 0x00, 0x00]
    ]
    
    i = 0
    for c in range(4):
        for r in range(4):
            ans[r][c] = int(t[i:i+2], 16)
            i += 2
    return ans

def convertKey128(k):
    ans = []
    
    i = 0
    while i < 32:
        ans.append(int(k[i:i+2], 16))
        i += 2
                   
    return ans

def convertKey192(k):
    ans = []
    
    i = 0
    while i < 48:
        ans.append(int(k[i:i+2], 16))
        i += 2
                   
    return ans

def convertKey256(k):
    ans = []
    
    i = 0
    while i < 64:
        ans.append(int(k[i:i+2], 16))
        i += 2
                   
    return ans
```
